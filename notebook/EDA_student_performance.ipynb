```python
# Numercial Features
num_features=[feature for feature in df.columns if df[feature].dtype!="O"]
cat_features=[feature for feature in df.columns if df[feature].dtype=="O"]
print(num_features)
print(cat_features)
```

```python
df["total_score"]=df["math_score"]+df["reading_score"]+df["writing_score"]
df["average_score"]=df["total_score"]/3
```

```python
reading_full=df[df["reading_score"]==100]["average_score"].count()
print(reading_full)
wriyting_full=df[df["writing_score"]==100]["average_score"].count()
print(wriyting_full)
math_full=df[df["math_score"]==100]["average_score"].count()
print(math_full)
```

```python
reading_less=df[df["reading_score"]<=20]["average_score"].count()
writing_less=df[df["writing_score"]<=20]["average_score"].count()
math_less=df[df["math_score"]<=20]["average_score"].count()
print(reading_less)
print(writing_less)
print(math_less)
```
